```python
import itertools
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split

# copia del dataset iniziale
new_data = dataset.copy()

# sostituisco i valori di tipo stringa nel nuovo dataset
new_data["Gender"] = new_data["Gender"].replace({"Male":0, "Female":1})
new_data["family_history_with_overweight"] = new_data["family_history_with_overweight"].replace({"no":0, "yes":1})
new_data["FAVC"] = new_data["FAVC"].replace({"no":0, "yes":1})
new_data["CAEC"] = new_data["CAEC"].replace({"no":0, "Sometimes":1, "Frequently":2, "Always":3})
new_data["SMOKE"] = new_data["SMOKE"].replace({"no":0, "yes":1})
new_data["SCC"] = new_data["SCC"].replace({"no":0, "yes":1})
new_data["CALC"] = new_data["CALC"].replace({"no":0, "Sometimes":1, "Frequently":2, "Always":3})
new_data["MTRANS"] = new_data["MTRANS"].replace({"Public_Transportation":0, "Walking":1, "Automobile":3, "Bike":4})
new_data["NObeyesdad"] = new_data["NObeyesdad"].replace({"Normal_Weight":0, "Overweight_Level_I":0, "Overweight_Level_II":0, "Insufficient_Weight":0, "Obesity_Type_I":1, "Obesity_Type_II":1, "Obesity_Type_III":1})

# randomizzo le righe del nuovo dataset
new_data = new_data.sample(frac = 1).reset_index(drop = True)

# funzione di normalizzazione
def normalize_dataset(data):
    norm_data = pd.DataFrame()

    for col in data.columns:
        media = data[col].mean()
        devstd = data[col].std()
        norm_data[col] = (data[col] - media) / devstd
        if devstd == 0:
            norm_data[col] = data[col]

    return norm_data

# divisione del dataset in target e gli altri attributi (per regressione)
x_reg = new_data.drop(columns = ['Weight'])
y_reg = new_data['Weight']

# divisione del dataset in target e gli altri attributi (per classificazione)
x_cla = new_data.drop(columns = ['NObeyesdad'])
y_cla = new_data['NObeyesdad']

# converto le colonne in tipo numerico
x_reg = x_reg.apply(pd.to_numeric, errors = 'coerce')
y_reg = pd.to_numeric(y_reg, errors = 'coerce')
x_cla = x_cla.apply(pd.to_numeric, errors = 'coerce')
y_cla = pd.to_numeric(y_cla, errors = 'coerce')

# rimuovo i valori NaN
x_reg = x_reg.dropna()
x_cla = x_cla.dropna()
y_reg = y_reg.loc[x_reg.index]
y_cla = y_cla.loc[x_cla.index]

# eseguo la normalizzazione
x_reg = normalize_dataset(x_reg)
x_cla_norm = normalize_dataset(x_cla)

# inserisco la colonna di bias per la regressione lineare
x_reg.insert(0, 'bias', 1)

# divisione del dataset in training set (~90%) e test set (~10%)
x_reg_train_bias, x_reg_test_bias, y_reg_train, y_reg_test = train_test_split(x_reg, y_reg, test_size = 0.1)
x_cla_train, x_cla_test, y_cla_train, y_cla_test = train_test_split(x_cla, y_cla, test_size = 0.1)
x_cla_train_norm, x_cla_test_norm, y_cla_train_norm, y_cla_test_norm = train_test_split(x_cla_norm, y_cla, test_size = 0.1)
```

# Linear Regression

**Implementazione di algoritmi di regressione lineare al fine di prevedere il valore Weight**

### 1. Linear regression con risoluzione basata su matrice:
Input:
1. X -> DataFrame
2. y -> DataFrame

Output:
1. β -> array

Il suo funzionamento si basa sulla classica formula di Linear Regression matrice in cui si moltiplica la pseudoinversa di X per il vettore y (facendo uso delle librerie numpy), ottenendo i coefficienti di regressione β.

### 2. Linear regression con risoluzione basata su Mini-Batch Stochatic Gradient Descent:
Input:
1. X -> DataFrame
2. y -> DataFrame
3. learning_rate -> int
4. epochs -> int
5. batch_size -> int

Output:
1. w -> array

La sgd() ha il fine di implementare una gradient discent con batch_size variabile, al fine di implementare diversi tipi di Stochastic Gradient Discent (tra le quali Mini-Batch di cui parlerò in seguito). Attraverso più epoche (ciclo for) vengono aggiornati linearmente i pesi del modello su piccoli batch di dati (ciclo for).

Per questo Homework, ho deciso di implementare una Mini-Batch Stochastic Gradient Descend, settando nella scelta degli iperparametri, il batch_size rispettivamente a 16, 32 e 64 piuttosto che a 1 come le normali SGD. Questa scelta è dovuta a un forte incremento di performance temporale in quantità di tempo non andando a discriminare la qualità della regressione per questo specifico dataset.

*Per quanto riguarda il tuning degli iperparametri, ho usato una classica Grid Search, di cui parlarò in seguito in un paragrafo dedicato.*



```python
from sklearn.linear_model import LinearRegression

# funzione di regressione lineare (come definita nelle slides)
def linear_regression(data, res):
    x_m = data.values
    y_v = res.values
    c = np.dot(x_m.T, x_m)
    c = np.linalg.inv(c)
    c = np.dot(c, x_m.T)
    c = np.dot(c, y_v)
    return c

# funzione di predizione
def predict_linear(test, weight):
    x_m_test = test
    y_pr = np.dot(x_m_test, weight)
    return y_pr

# allenamento del modello
weights = linear_regression(x_reg_train_bias, y_reg_train)

# predizione sul test set
predictions = predict_linear(x_reg_test_bias, weights)

# calcolo RMSE
rmse = np.sqrt(mean_squared_error(predictions, y_reg_test))
print(f'Linear Regression - RMSE: {rmse}')

# regressione con sklearn
model_sklearn_lin = LinearRegression()
model_sklearn_lin.fit(x_reg_train_bias, y_reg_train)

predictions_sklearn_lin = model_sklearn_lin.predict(x_reg_test_bias)
rmse_sklearn = np.sqrt(mean_squared_error(predictions_sklearn_lin, y_reg_test))
print(f'Sklearn Linear Regression - RMSE: {rmse_sklearn}')
print('-----------------')

#-----------------Gradient Descent (Mini-Batch)-----------------

# setto un seme di randomizzazione globale
np.random.seed(42)

# funzione per la regressione lineare usando Mini-Batch derivata dall'uso specifico della Stochastic Gradient Descent
def sgd(X, y, lr, epochs, batch_size):
    w = np.zeros(X.shape[1])
    n = len(y)
    np.random.seed(42)

    for epoch in range(epochs):
        indices = np.arange(n)
        np.random.shuffle(indices)
        X = X.iloc[indices].reset_index(drop = True)
        y = y.iloc[indices].reset_index(drop = True)

        for start in range(0, n, batch_size):
            end = start + batch_size
            X_batch = X.iloc[start:end]
            y_batch = y.iloc[start:end]
            prediction = np.dot(X_batch, w)
            error = prediction - y_batch
            gradient = np.dot(X_batch.T, error) / batch_size
            gradient = np.clip(gradient, -1e3, 1e3)
            w -= lr * gradient

    return w

def predict_sgd(X, w):
    return np.dot(X, w)

# funzione di grid search per sgd
def grid_search(train_function, predict_function, h_params, train, res, validation, validation_res):
    best_score = None
    chosen_params = None

    # generare tutte le possibili combinazioni di iperparametri dal dizionario
    h_params_comb = list(itertools.product(*h_params.values()))

    for params in h_params_comb:
        # chiamare le funzioni di train e predict con l'insieme di iperparametri corrente
        w = train_function(train, res, *params)

        pred = predict_function(validation, w)

        # stampare i parametri per questa iterazione
        for i, key in enumerate(h_params.keys()):
            print(key,': ', params[i],end='; ',sep='')

        # calcolare l'accuratezza o il MSE
        print('RMSE:',end=' ')
        score = np.sqrt(mean_squared_error(validation_res, pred))

        if best_score is None or score < best_score:
            best_score = score
            chosen_params = params

        print(score)
    return chosen_params

# parametri per grid search
h_params = {
    'lr': [0.001, 0.01, 0.1],
    'epochs': [3, 5, 7],
    'batch_size': [16, 32, 64] # mini-batch (modicare questo valore per avere diversi tipi di gradient descent)
}

# esecuzione grid search
best_params_sgd = grid_search(sgd, predict_sgd, h_params, pd.DataFrame(x_reg_train_bias), pd.Series(y_reg_train), pd.DataFrame(x_reg_test_bias), pd.Series(y_reg_test))
print("Best hyperparameters:", best_params_sgd)

# addestro il modello con gli iperparametri scelti
final_w_sgd = sgd(x_reg_train_bias, y_reg_train, *best_params_sgd)
y_reg_pred_sgd = predict_sgd(x_reg_test_bias, final_w_sgd)

rmse_sgd = np.sqrt(mean_squared_error(y_reg_test, y_reg_pred_sgd))
print("Mini-Batch - RMSE:", rmse_sgd)
```

# Decision Tree
**Implementazione dell'albero di decisione al fine di prevedere il valore NObeyesdad**

Per quest'algoritmo viene usata l'entropia per misurare l'impurità dei nodi e il guadagno informativo per la scelta della miglior caratteristica per la suddivisione. Quindi viene costruito ricorsivamente l'albero, in base alla caratteristica scelta e suddividendo i dati fino al raggiungimento della profondità massima o all'ottenimento di nodi puri

### 1. info_gain()
Input:
1. X -> DataFrame
2. y -> DataFrame
3. index

Output:
1. gain

Calcola l'entropia di base. Trovati quindi i valori unici, calcola l'entropia del sottoinsieme e somma l'entropia pesata. Quindi sottrae l'entropia pesata.

### 2. Node
Classe che rappresenta i nodi dell'albero

### 3. build_tree()
Input:
1. X -> DataFrame
2. y -> DataFrame
3. depht -> int
4. max_depht -> int

Output:
1. node -> Node

Se tutti i campioni ottenuti sono della stessa classe o max_depht, crea un nodo figlio. Quindi calcola l'information gain di ogni caratteristica, scegliendo qualle con info_gain massimo e suddividendo i dati in due sottoinsieme a seconda di ciò. Quindi costruisce ricorsivamente il sottoalbero sx e dx, restituendo alla fine un nodo con: la caratteristica scelta, soglia e sottoalberi sx e dx.

### 4. predict_tree()
Input:
1. node -> Node
2. x -> DataFrame

Output:
1. classe

Se il nodo è una foglia, ne restituisce il valore. Se la caratteristica è inferiore alla soglia del punto, chiama ricorsivamente predict_tree() sul sottoalbero sx, altrimenti su quello dx.

*Il Decision Tree qua implementato non esegue tuning di iperparametri specifici come ad esempio l'altezza dell'albero, al fine di semplificarne l'addestramento e l'implementazione, nonché di dare liberà di adattamento ai dati, tuttavia incorrendo ad un eventuale rischio di overfitting*



```python
from sklearn.tree import DecisionTreeClassifier

# funzione per il calcolo dell'entropia
def entropy(y):
    class_count = np.bincount(y)
    probabilities = class_count / len(y)
    return -np.sum([p * np.log2(p) for p in probabilities if p > 0])

# funzione per il calcolo del guadagno informativo
def info_gain(X, y, index):
    base_entropy = entropy(y)
    unique_values = np.unique(X[:, index])
    weighted_entropy = 0

    for value in unique_values:
        sub_y = y[X[:, index] == value]
        weighted_entropy += (len(sub_y) / len(y)) * entropy(sub_y)

    return base_entropy - weighted_entropy

# classe del nodo dell'albero
class Node:
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, value=None):
        self.feature_index = feature_index
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

# funzione per costruire l'albero
def build_tree(X, y, depth = 0, max_depth = 7):
    n_samples, n_features = X.shape
    unique_classes = np.unique(y)

    if len(unique_classes) == 1 or depth >= max_depth:
        leaf_value = unique_classes[0]
        return Node(value=leaf_value)

    best_gain = 0
    best_feature = None
    best_threshold = None

    for feature_index in range(n_features):
        gain = info_gain(X, y, feature_index)
        if gain > best_gain:
            best_gain = gain
            best_feature = feature_index
            best_threshold = 0.5  # soglia binaria 0.5 per le caratteristiche normalizzate

    if best_gain == 0:
        leaf_value = unique_classes[np.argmax(np.bincount(y))]
        return Node(value=leaf_value)

    left_indices = X[:, best_feature] < best_threshold
    right_indices = X[:, best_feature] >= best_threshold

    if len(y[left_indices]) == 0 or len(y[right_indices]) == 0:
        leaf_value = unique_classes[np.argmax(np.bincount(y))]
        return Node(value=leaf_value)

    left_subtree = build_tree(X[left_indices], y[left_indices], depth + 1, max_depth)
    right_subtree = build_tree(X[right_indices], y[right_indices], depth + 1, max_depth)
    return Node(feature_index=best_feature, threshold=best_threshold, left=left_subtree, right=right_subtree)

def predict_tree(node, X):
    if node.value is not None:
        return node.value
    if X[node.feature_index] < node.threshold:
        return predict_tree(node.left, X)
    else:
        return predict_tree(node.right, X)

# addestro il modello
tree = build_tree(x_cla_train_norm.values, y_cla_train_norm.values, max_depth = 7)
predictions_t = [predict_tree(tree, sample) for sample in x_cla_test_norm.values]
accuracy_tree = accuracy_score(y_cla_test_norm, predictions_t)
print(f'Decision Tree - Accuracy: {accuracy_tree * 100}')

# addestro il modello sklearn
clf = DecisionTreeClassifier(max_depth = 7)
clf.fit(x_cla_train_norm, y_cla_train_norm)
predictions_tree_sklearn = clf.predict(x_cla_test_norm)
accuracy_tree_sklearn = accuracy_score(y_cla_test_norm, predictions_tree_sklearn)
print(f'Sklearn Decision Tree - Accuracy: {accuracy_tree_sklearn * 100}')
```

#Logistic Regression
**Implementazione dell'algoritmo di regressione logistica al fine di prevedere il valore NObeyesdad**

Input:
1. X -> DataFrame
2. y -> DataFrame
3. learning_rate -> int
4. iterations -> int

Output:
1. weights -> array
2. accuracy_progress -> list

In questa implementazione di Logistic Regression, viene usata la Gradient Descent per aggiornare linearmente il peso del modello. Dopo aver inizializzato i pesi a zero infatti, la funzione, per ogni iterazione:
- calcola le probabilità predette tramite sigmoid()
- calcola il gradiente
- aggiorno i pesi in base al gradiente e al learning_rate
- memorizza l'accuratezza del modello (per il grafico)

*Per quanto riguarda il tuning degli iperparametri, ho usato una classica Grid Search, di cui parlarò in seguito in un paragrafo dedicato.*


```python
from sklearn.linear_model import LogisticRegression

# funzione di grid search per logistic regression
def grid_search_log(train_function, predict_function, h_params, train, res, validation, validation_res):
    best_score = None
    chosen_params = None
    best_accuracy_progress = None
    accuracy_results = []

    # genero tutte le possibili combinazioni di iperparametri
    h_params_comb = list(itertools.product(*h_params.values()))

    for params in h_params_comb:
        # addestro il modello con il set corrente di iperparametri
        w, accuracy_progress = train_function(train, res, *params)

        # predizioni sul set di validazione
        pred = predict_function(validation, w)

        # stampo gli iperparametri correnti
        for i, key in enumerate(h_params.keys()):
            print(key, ': ', params[i], end='; ', sep='')

        print('Accuracy:', end=' ')
        score = accuracy_score(validation_res, pred) * 100
        accuracy_results.append((params[0], score))

        if best_score is None or score > best_score:
            best_score = score
            chosen_params = params
            best_accuracy_progress = accuracy_progress

        print(score)

     # calcolare l'accuratezza media per ogni learning_rate
    accuracy_per_lr = {}
    for lr, score in accuracy_results:
        if lr not in accuracy_per_lr:
            accuracy_per_lr[lr] = []
        accuracy_per_lr[lr].append(score)

    avg_accuracy_per_lr = {lr: np.mean(scores) for lr, scores in accuracy_per_lr.items()}

    return chosen_params, best_accuracy_progress, avg_accuracy_per_lr

# funzione sigmoid
def sigmoid(x):
    x = np.clip(x, -500, 500) # per evitare overflow
    return 1 / (1 + np.exp(-x))

def logistic_regression(X, y, learning_rate, iterations):
    weights = np.zeros(X.shape[1])
    m = len(y)
    accuracy_progress = []

    # ciclo di iterazioni per aggiornare i pesi
    for _ in range(iterations):
        predictions = sigmoid(np.dot(X, weights))
        gradient = np.dot(X.T, (predictions - y)) / m
        weights -= learning_rate * gradient

        pred = predict_logistic(X, weights)
        acc = accuracy_score(y, pred) * 100
        accuracy_progress.append(acc)
    return weights, accuracy_progress

def predict_logistic(X, weights):
    return sigmoid(np.dot(X,weights)) >= 0.5

# definizione degli iperparametri da cercare
h_params = {
    'Learning_Rate': [0.0001, 0.001, 0.01, 0.1],
    'Iterations': [200, 400, 600]
}

# esecuzione grid search
best_params, accuracy_progress, avg_accuracy_per_lr = grid_search_log(logistic_regression, predict_logistic, h_params, x_cla_train_norm, y_cla_train_norm, x_cla_test_norm, y_cla_test_norm)
print('Best Hyperparameters: ', best_params)

# estrazione migliori iper-parametri
best_learning_rate, best_iterations = best_params

# modello finale con migliori iperparametri
final_weights, final_accuracy_progress = logistic_regression(x_cla_train_norm, y_cla_train_norm, best_learning_rate, best_iterations)

# previsioni sul test set
y_pred_test = predict_logistic(x_cla_test_norm, final_weights)

# calcolo dell'accuratezza
final_accuracy = accuracy_score(y_cla_test_norm, y_pred_test) * 100
print(f'Logistic Regression - Accuracy: {final_accuracy}')

# regressione logistica di sklearn
model_sklearn_log = LogisticRegression(max_iter=600)
model_sklearn_log.fit(x_cla_train_norm, y_cla_train_norm)

# previsioni sul test con sklearn
y_pred_test_sklearn = model_sklearn_log.predict(x_cla_test_norm)

# calcolo dell'accuratezza sklearn (per confronto)
final_accuracy_log_sklearn = accuracy_score(y_cla_test_norm, y_pred_test_sklearn) * 100
print(f'Sklearn Logistic Regression - Accuracy: {final_accuracy_log_sklearn}')

# grafico sulle prestazioni
learning_rates = sorted(avg_accuracy_per_lr.keys())
accuracies = [avg_accuracy_per_lr[lr] for lr in learning_rates]

plt.plot(learning_rates, accuracies, marker='o', label='Accuracy')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy in Logistic Regression')
plt.xscale('log')  # scala logaritmica
plt.legend()
plt.show()
```

# K-Nearest Neighbors (KNN)
**Implementazione dell'algoritmo di KNN al fine di prevedere il valore NObeyesdad**

Esso trova i k punti più vicini al punto in analisi e gli assegna la classe più frequente tra i vicini. Il suo funzionamento avviene attraverso diverse funzioni: viene prima chiamata la predict_knn che itera su ogni punto di test; per ogni punto di test viene chiamata la knn per trovare i k più vicini e subito dopo la knn_mv per determinarne la classe più comune.

### 1. knn()
Input:
1. X -> DataFrame
2. y -> DataFrame
3. test_point
4. k -> int

Output:
1. nearest_neighbors -> array

Converte X, y e test_point in array numpy, calcola la distanza euclidea rispetto ai punti di addestramento, ordine le distanze in ordine crescente e restituisce le etichette dei k più vicini

### 2. knn_mv()
Input:
1. nearest_neighbors -> array

Output:
1. classe più frequentata

tramite np.bincount conta le occorrenze di ogni etichetta e restituisce quella con valore più alto

### 3. predict_knn()
Input:
1. X -> DataFrame
2. y -> DataFrame
3. test_point
4. k -> int

Output:
1. predictions -> array

Itera su ogni punto di test per il quale chiama knn() e knn_mv(), aggiungendo la classe predetta e restituendo alla fine l'array delle predizioni

*Per quanto riguarda il tuning degli iperparametri, ho usato una classica Grid Search, di cui parlarò in seguito in un paragrafo dedicato.*

```python
from sklearn.neighbors import KNeighborsClassifier

# funzione di grid search per KNN
def grid_search_knn(predict_function, h_params, train, res, validation, validation_res):
    best_score = None
    chosen_param = None
    accuracy_results = []

    h_params_comb = list(itertools.product(*h_params.values()))

    for params in h_params_comb:
        k = params[0]
        pred = predict_function(train, res, validation, k)

        for i, key in enumerate(h_params.keys()):
            print(key, ': ', params[i], end='; ', sep='')

        print('Accuracy:', end=' ')
        score = accuracy_score(validation_res, pred) * 100
        accuracy_results.append((k, score))
        if best_score is None or score > best_score:
            best_score = score
            chosen_param = k
        print(score)
    return chosen_param, accuracy_results

# funzioni per KNN
def knn(train_data, train_labels, test_point, k):
    train_data_np = train_data.to_numpy()
    train_labels_np = train_labels.to_numpy()
    test_point_np = test_point.to_numpy().reshape(1, -1)

    distances = np.sqrt(((train_data_np - test_point_np) ** 2).sum(axis=1))
    sorted_indices = np.argsort(distances)[:k]
    nearest_neighbors = train_labels_np[sorted_indices]
    return nearest_neighbors

def knn_mv(nearest_neighbors):
    return np.bincount(nearest_neighbors).argmax()

def predict_knn(train_data, train_labels, test_data, k):
    predictions = []
    for i in range(len(test_data)):
        nearest_neighbors = knn(train_data, train_labels, test_data.iloc[i], k)
        predictions.append(knn_mv(nearest_neighbors))
    return np.array(predictions)

# parametri per la grid search
h_params = {
    'k': list(range(1, 31))
}

# esecuzione della grid search per trovare gli iper-parametri migliori per classificazione (NObeyesdad)
best_k, accuracy_results = grid_search_knn(predict_knn, h_params, x_cla_train, y_cla_train, x_cla_test, y_cla_test)
print('Best Hyperparameter:', best_k)

# addestramento modello
knn = predict_knn(x_cla_train, y_cla_train, x_cla_test, best_k)
accuracy_knn = accuracy_score(y_cla_test, knn) * 100
print(f'KNN Classification - Accuracy: {accuracy_knn}')

# knn classificazion con sklearn
model_clf = KNeighborsClassifier(n_neighbors = best_k)
model_clf.fit(x_cla_train, y_cla_train)
y_pred_clf = model_clf.predict(x_cla_test)
accuracy_knn = accuracy_score(y_cla_test, y_pred_clf) * 100
print(f'Sklearn KNN Classification - Accuracy: {accuracy_knn}')

# grafico sulle prestazioni
ks = [result[0] for result in accuracy_results]
accuracies = [result[1] for result in accuracy_results]

plt.plot(ks, accuracies, marker='o', label='Accuracy')
plt.xlabel('k')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy in KNN')
plt.legend()
plt.grid(True)
plt.show()
```

# Neural Network
**Implementazione dell'algoritmo di rete neurale al fine di prevedere il valore Weight**

La rete neurale feed_forward implementata ha un'architettura a più strati:
- input: dimensione determinata dal numero di features nel DatFrame
- output: in questo caso a singola unità visto la singolarità del target
- nascosti: definito dal 'k'

### 1. forward_prop()
Input:
1. test
2. weights
3. bias

Output:
1. list

Calcola le attivazioni per i vari strati applicando agli strati nascosti la funzione sigmoid(), già discussa nel merito di Logistic Regression

### 2. back_prop()
Input:
1. res
2. weights
3. bias
4. list
5. learning rate

Output:
1. weights
2. bias

Calcola i gradienti della funzione di perdita rispetto ai pesi e ai bias usando l'algoritmo di backpropagation e aggiorna i pesi e i bias di conseguenza

### 3. neural_network()
Input:
1. data
2. res
3. k
4. learning rate
5. epochs
6. hidden_levels

Output:
1. weights
2. bias
3. mse_list

Esegue backpropagation e forward propagation a seconda del numero di epoche, registrando RMSE in ogni epoca (per il grafico prestazionale)

### 4. predict_nn()
Input:
1. test
2. weights
3. bias

Output:
1. predizioni

Esegue forward propagation sui dati di test

*Per quanto riguarda il tuning degli iperparametri, ho usato una classica Grid Search, di cui parlarò in seguito in un paragrafo dedicato.*

```python
from sklearn.linear_model import LinearRegression

# setto un seme di randomizzazione globale
np.random.seed(42)

# funzione sigmoid
# già vista nella logistic regression e reinserita in questo codice per garantirne l'indipendenza di esecuzione
def sigmoid(x):
    x = np.clip(x, -500, 500) # per evitare overflow
    return 1 / (1 + np.exp(-x))

def forward_prop(data, weights, biases):
    output = [data]

    for i in range(len(weights) - 1):
        bi_mat = np.tile(biases[i], (output[i].shape[0], 1))
        out = sigmoid(np.dot(output[i], weights[i]) + bi_mat)
        output.append(out)

    b_last = np.tile(biases[-1], (output[-1].shape[0], 1))
    out = np.dot(output[-1], weights[-1]) + b_last
    output.append(out)
    return output

def back_prop(res, weights, biases, outputs, lr):
    alpha = lr

    for i in reversed(range(len(weights))):
        y_pred = outputs[i + 1]
        a_mat = outputs[i]
        if i == len(weights) - 1:
            delta = (res - y_pred)
        else:
            sigm_der = y_pred * (1 - y_pred)
            delta = np.dot(delta, weights[i + 1].T) * sigm_der
        weights[i] += alpha * (1 / a_mat.shape[0]) * np.dot(a_mat.T, delta)
        delta_sum = np.sum(delta, axis=0)
        delta_sum = delta_sum.reshape(1, -1)
        biases[i] += alpha * (1 / a_mat.shape[0]) * delta_sum

    return weights, biases

def neural_network(data, res, k = 5, lr = 0.001, epochs = 200, hidden_levels = 1):
    np.random.seed(42)
    mse_list = []
    w = []
    b = []
    w.append(np.random.randn(data.shape[1], k))
    b.append(np.random.randn(1, k))

    for i in range(hidden_levels - 1):
        w.append(np.random.randn(k, k))
        b.append(np.random.randn(1, k))

    w.append(np.random.randn(k).reshape(-1, 1))
    b.append(np.random.randn(1).reshape(-1, 1))

    for i in range(epochs):
        output = forward_prop(data, w, b)
        w, b = back_prop(res, w, b, output, lr)
        mse = mean_squared_error(res, output[-1])
        mse_list.append(mse)

    return w, b, mse_list

def predict_nn(test, w, b):
    return forward_prop(test, w, b)[-1]

# grid search per neural network
def grid_search_nn(train_function, predict_function, h_params, train, res, validation, validation_res):
    best_score = None
    chosen_params = None
    k_rmse_results = []

    # genero tutte le possibili combinazioni di iperparametri
    h_params_comb = list(itertools.product(*h_params.values()))

    for params in h_params_comb:
        # addestro il modello con il set corrente di iperparametri
        w, b, mse_list = train_function(train, res, *params)

        # predizioni sul set di validazione
        pred = predict_function(validation, w, b)

        # stampo gli iperparametri correnti
        for i, key in enumerate(h_params.keys()):
            print(key, ': ', params[i], end='; ', sep='')

        print(' RMSE:', end=' ')
        score = np.sqrt(mean_squared_error(pred, validation_res))
        k = params[0]
        k_rmse_results.append((k, score))

        # calcolo l'accuratezza
        if best_score is None or score < best_score:
            best_score = score
            chosen_params = params
        print(score)
    return chosen_params, k_rmse_results

h_params = {
    'k': [3, 5, 7],
    'lr': [0.001, 0.01, 0.1],
    'epochs': [200, 500, 800],
    'hidden_levels': [1] # un solo livello come richiesto (modificare questo parametro per avere più livelli nascosti)
}

# cerco i migliori iperparametri con grid search
best_params, k_rmse_results = grid_search_nn(neural_network, predict_nn, h_params, x_reg_train_bias, y_reg_train.values.reshape(-1, 1), x_reg_test_bias, y_reg_test)
print("Best Hyperparameters:", best_params)

# addestro il neural network
weights, biases, mse_list = neural_network(x_reg_train_bias, y_reg_train.values.reshape(-1, 1), *best_params)
y_reg_test_pred_nn = predict_nn(x_reg_test_bias, weights, biases)
rmse_nn = np.sqrt(mean_squared_error(y_reg_test, y_reg_test_pred_nn))
print("Neural Network - RMSE:", rmse_nn)

# linear regression con sklearn per confronto performance
lin_reg = LinearRegression()
lin_reg.fit(x_reg_train_bias, y_reg_train)
y_reg_test_pred_lin = lin_reg.predict(x_reg_test_bias)
rmse_lin = np.sqrt(mean_squared_error(y_reg_test, y_reg_test_pred_lin))
print("Sklearn Linear Regression - RMSE:", rmse_lin)

# grafico sulle prestazioni
ks = [result[0] for result in k_rmse_results]
rmses = [result[1] for result in k_rmse_results]

plt.stem(ks, rmses)
plt.xlabel('k')
plt.ylabel('RMSE')
plt.title('RMSE in Neural Network')
plt.grid(True)
plt.show()
```
