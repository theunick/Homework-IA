# Data preparation
**Come prima cosa si sostituiscono i valori testuali nel dataset con valori numerici. Dopo aver randomizzato le righe del nuovo dataset, i valori vengono quindi normalizzati per ridimensionare i dati. Inoltre si divide anche il nuovo dataset aggiornato in training set e test set con un rapporto di circa 90/10.**

> Legenda del nuovo dataset:
* Gender -> 0 male, 1 female
* FHWO / FAVC / Smoke / SCC -> 0 no, 1 yes
* CEAC / CALC -> 0 no, 1 sometimes, 2 frequently, 3 always
* MTRANS -> 0 public_trans, 1 walking, 2 automobile, 3 motorbike, 4 bike
* NObeyesdad -> 0 normal / overw_L1 / overw_L2 / insuff, 1 obesity_L1 / obesity_L2 / obesity_L3

> Normalizzazione applicata per ridimensionare i dati. Ogni colonna è stata trasformata, garantendo quindi che tutte le caratteristiche abbiano media zero e deviazione standard unitaria, riducendo il rischio che alcune variabili dominino il modello a causa della loro scala.

> Creazione dei rispettivi dataset:
* reg -> normalizzato e con il target e le feautures tarate per la regressione sui valori Weight
* cla -> **non** normalizzato e con il target e le feautures tarate per la classificazione sui valori NObeyesdad
* cla_norm -> normalizzato e con il target e le feautures tarate per la classificazione sui valori NObeyesdad

> Divisione dei dataset in training set ~90% e test set ~10%

```python
import itertools
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split

# copia del dataset iniziale
new_data = dataset.copy()

# sostituisco i valori di tipo stringa nel nuovo dataset
new_data["Gender"] = new_data["Gender"].replace({"Male":0, "Female":1})
new_data["family_history_with_overweight"] = new_data["family_history_with_overweight"].replace({"no":0, "yes":1})
new_data["FAVC"] = new_data["FAVC"].replace({"no":0, "yes":1})
new_data["CAEC"] = new_data["CAEC"].replace({"no":0, "Sometimes":1, "Frequently":2, "Always":3})
new_data["SMOKE"] = new_data["SMOKE"].replace({"no":0, "yes":1})
new_data["SCC"] = new_data["SCC"].replace({"no":0, "yes":1})
new_data["CALC"] = new_data["CALC"].replace({"no":0, "Sometimes":1, "Frequently":2, "Always":3})
new_data["MTRANS"] = new_data["MTRANS"].replace({"Public_Transportation":0, "Walking":1, "Automobile":3, "Bike":4})
new_data["NObeyesdad"] = new_data["NObeyesdad"].replace({"Normal_Weight":0, "Overweight_Level_I":0, "Overweight_Level_II":0, "Insufficient_Weight":0, "Obesity_Type_I":1, "Obesity_Type_II":1, "Obesity_Type_III":1})

# randomizzo le righe del nuovo dataset
new_data = new_data.sample(frac=1).reset_index(drop=True)

# funzione di normalizzazione
def normalize_dataset(data):
    norm_data = pd.DataFrame()

    for col in data.columns:
        media = data[col].mean()
        devstd = data[col].std()
        norm_data[col] = (data[col] - media) / devstd
        if devstd == 0:
            norm_data[col] = data[col]

    return norm_data

# divisione del dataset in target e gli altri attributi (per regressione)
x_reg = new_data.drop(columns = ['Weight'])
y_reg = new_data['Weight']

# divisione del dataset in target e gli altri attributi (per classificazione)
x_cla = new_data.drop(columns = ['NObeyesdad'])
y_cla = new_data['NObeyesdad']

# converto le colonne in tipo numerico
x_reg = x_reg.apply(pd.to_numeric, errors = 'coerce')
y_reg = pd.to_numeric(y_reg, errors = 'coerce')
x_cla = x_cla.apply(pd.to_numeric, errors = 'coerce')
y_cla = pd.to_numeric(y_cla, errors = 'coerce')

# rimuovo i valori NaN
x_reg = x_reg.dropna()
x_cla = x_cla.dropna()
y_reg = y_reg.loc[x_reg.index]
y_cla = y_cla.loc[x_cla.index]

# eseguo la normalizzazione
x_reg = normalize_dataset(x_reg)
x_cla_norm = normalize_dataset(x_cla)

# inserisco la colonna di bias per la regressione lineare
x_reg.insert(0, 'bias', 1)

# divisione del dataset in training set (~90%) e test set (~10%)
x_reg_train_bias, x_reg_test_bias, y_reg_train, y_reg_test = train_test_split(x_reg, y_reg, test_size = 0.1)
x_cla_train, x_cla_test, y_cla_train, y_cla_test = train_test_split(x_cla, y_cla, test_size = 0.1)
x_cla_train_norm, x_cla_test_norm, y_cla_train_norm, y_cla_test_norm = train_test_split(x_cla_norm, y_cla, test_size = 0.1)
```

# Linear regression

**Implementazione dell'algoritmo di regressione lineare al fine di prevedere il valore Weight**

> Funzioni sviluppate:
* funzione di regressione lineare -> come sulle slides
* funzione di predizione (standard)

Considerazioni dell'algortimo:


```python
from sklearn.linear_model import LinearRegression

# funzione di regressione lineare (come definita nelle slides)
def linear_regression(data, res):
    x_m = data.values
    y_v = res.values
    c = np.dot(x_m.T, x_m)
    c = np.linalg.inv(c)
    c = np.dot(c, x_m.T)
    c = np.dot(c, y_v)
    return c

# funzione di predizione
def predict_linear(test, weight):
    x_m_test = test
    y_pr = np.dot(x_m_test, weight)
    return y_pr

# allenamento del modello
weights = linear_regression(x_reg_train_bias, y_reg_train)

# predizione sul test set
predictions = predict_linear(x_reg_test_bias, weights)

# calcolo RMSE
rmse = np.sqrt(mean_squared_error(predictions, y_reg_test))
print(f'Linear Regression - RMSE: {rmse}')

# regressione con sklearn
model_sklearn_lin = LinearRegression()
model_sklearn_lin.fit(x_reg_train_bias, y_reg_train)

predictions_sklearn_lin = model_sklearn_lin.predict(x_reg_test_bias)
rmse_sklearn = np.sqrt(mean_squared_error(predictions_sklearn_lin, y_reg_test))
print(f'Sklearn Linear Regression - RMSE: {rmse_sklearn}')
```

#Logistic Regression
**Implementazione dell'algoritmo di regressione logistica al fine di prevedere il valore NObeyesdad**

> Funzioni sviluppate:
* grid search per il tuning degli iper-paramentri
* funzione di sigmoid
* funzione di regressione logistica
* funzione di predizione (standard)


```python
from sklearn.linear_model import LogisticRegression

# funzione di grid search per logistic regression
def grid_search_log(train_function, predict_function, h_params, train, res, validation, validation_res):
    best_score = None
    chosen_params = None

    # genero tutte le possibili combinazioni di iperparametri
    h_params_comb = list(itertools.product(*h_params.values()))

    for params in h_params_comb:
        # addestro il modello con il set corrente di iperparametri
        w = train_function(train, res, *params)

        # predizioni sul set di validazione
        if isinstance(w, tuple):
            pred = predict_function(validation, *w)
        else:
            pred = predict_function(validation, w)

        # stampo gli iperparametri correnti
        for i, key in enumerate(h_params.keys()):
            print(key, ': ', params[i], end='; ', sep='')

        # calcolo l'accuratezza
        print('Accuracy:', end=' ')
        score = accuracy_score(validation_res, pred) * 100

        if best_score is None or score > best_score:
            best_score = score
            chosen_params = params
        print(score)
    return chosen_params

# funzione sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logistic_regression(X, y, learning_rate, iterations):
    weights = np.zeros(X.shape[1])
    m = len(y)

    # ciclo di iterazioni per aggiornare i pesi
    for _ in range(iterations):
        predictions = sigmoid(np.dot(X, weights))
        gradient = np.dot(X.T, (predictions - y)) / m
        weights -= learning_rate * gradient
    return weights

def predict_logistic(X, weights):
    return sigmoid(np.dot(X,weights)) >= 0.5

# definizione degli iperparametri da cercare
h_params = {
    'Learning_Rate': [0.0001, 0.001, 0.01, 0.1],
    'Iterations': [200, 400, 600]
}

# esecuzione grid search
best_params = grid_search_log(logistic_regression, predict_logistic, h_params, x_cla_train_norm, y_cla_train_norm, x_cla_test_norm, y_cla_test_norm)
print('Best Hyperparameters: ', best_params)

# estrazione migliori iper-parametri
best_learning_rate, best_iterations = best_params

# modello finale con migliori iperparametri
final_weights = logistic_regression(x_cla_train_norm, y_cla_train_norm, best_learning_rate, best_iterations)

# previsioni sul test set
y_pred_test = predict_logistic(x_cla_test_norm, final_weights)

# calcolo dell'accuratezza
final_accuracy = accuracy_score(y_cla_test_norm, y_pred_test) * 100
print(f'Logistic Regression - Test Accuracy: {final_accuracy}')

# regressione logistica di sklearn
model_sklearn_log = LogisticRegression(max_iter=600)
model_sklearn_log.fit(x_cla_train_norm, y_cla_train_norm)

# previsioni sul test con sklearn
y_pred_test_sklearn = model_sklearn_log.predict(x_cla_test_norm)

# calcolo dell'accuratezza sklearn (per confronto)
final_accuracy_log_sklearn = accuracy_score(y_cla_test_norm, y_pred_test_sklearn) * 100
print(f'Sklearn Logistic Regression - Test Accuracy: {final_accuracy_log_sklearn}')
```

# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor

# funzione di grid search per KNN
def grid_search_knn(predict_function, h_params, train, res, validation, validation_res):
    best_score = None
    chosen_param = None

    h_params_comb = list(itertools.product(*h_params.values()))

    for params in h_params_comb:
        k = params[0]
        pred = predict_function(train, res, validation, k)

        for i, key in enumerate(h_params.keys()):
            print(key, ': ', params[i], end='; ', sep='')

        print('Accuracy:', end=' ')
        score = accuracy_score(validation_res, pred) * 100
        if best_score is None or score > best_score:
            best_score = score
            chosen_param = k
        print(score)
    return chosen_param

# funzioni per KNN
def knn(train_data, train_labels, test_point, k):
    train_data_np = train_data.to_numpy()
    train_labels_np = train_labels.to_numpy()
    test_point_np = test_point.to_numpy().reshape(1, -1)

    distances = np.sqrt(((train_data_np - test_point_np) ** 2).sum(axis=1))
    sorted_indices = np.argsort(distances)[:k]
    nearest_neighbors = train_labels_np[sorted_indices]
    return nearest_neighbors

def knn_mv(nearest_neighbors):
    return np.bincount(nearest_neighbors).argmax()

def predict_knn(train_data, train_labels, test_data, k, task):
    predictions = []
    for i in range(len(test_data)):
        nearest_neighbors = knn(train_data, train_labels, test_data.iloc[i], k)
        predictions.append(knn_mv(nearest_neighbors))
    return np.array(predictions)

def knn_classification(train_data, train_labels, test_data, k):
    return predict_knn(train_data, train_labels, test_data, k, 'classification')

# parametri per la grid search
h_params = {
    'k': [3, 5, 7, 9, 11, 15, 21, 101, 441]
}

# esecuzione della grid search per trovare gli iper-parametri migliori per classificazione (NObeyesdad)
best_k = grid_search_knn(knn_classification, h_params, x_cla_train, y_cla_train, x_cla_test, y_cla_test)
print('Best Hyperparameter for classification:', best_k)

# addestramento modello
final_knn = knn_classification(x_cla_train, y_cla_train, x_cla_test, best_k)
final_accuracy_knn = accuracy_score(y_cla_test, final_knn) * 100
print(f'Final KNN Classification Test Accuracy: {final_accuracy_knn}')

# knn classificazion con sklearn
model_clf = KNeighborsClassifier(n_neighbors = best_k)
model_clf.fit(x_cla_train, y_cla_train)
y_pred_clf = model_clf.predict(x_cla_test)
accuracy_knn = accuracy_score(y_cla_test, y_pred_clf) * 100
print(f'Sklearn KNN Classification Test Accuracy: {accuracy_knn}')
```
